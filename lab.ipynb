```python
import pyarrow.parquet as pq

# Carregar o arquivo Parquet
table = pq.read_table('data/TAB_CADASTRO_AL.parquet')

# Verificar informações básicas das colunas
print("Número de colunas:", table.num_columns)
print("Número de linhas:", table.num_rows)

# Listar todas as colunas e seus tipos
print("\nColunas e tipos:")
for i, field in enumerate(table.schema):
    print(f"{i+1}. {field.name}: {field.type}")

# Verificar schema completo
print("\nSchema completo:")
print(table.schema)
```
